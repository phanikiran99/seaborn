```python
# Lets import the modules 
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# create some linespace data
x = np.linspace(1,10,1000)  #Imports 1000 entries between 1 and 10 including 
plt.figure(figsize=(10,5))
plt.plot(x,np.sin(x),x,np.cos(x),label='Sin & Cos')  #Plots both sin and cos values for x 
plt.title('Sin and Cos Maps')
plt.xlabel('x')
plt.ylabel('sin/cos')
plt.legend()
```

```python
sns.set()
plt.figure(figsize=(10,5))
plt.plot(x,np.sin(x),x,np.cos(x),label='Sin & Cos')
plt.title('Sin and Cos Maps')
plt.xlabel('x')
plt.ylabel('sin/cos')
plt.legend()
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
```

```python
# before visualizing data it is better to understand it first 
data.describe()
```

```python
# Plotting Histograms 
# Histograms are generally plotted to a diagram consisting of rectangles 
# whose area is proportional to the frequency of a variable and whose width is equal to the class interval.
# for ex: population of multiple countries(x ctry, y population), rainfall in an year etc.. (x year, y rain fall in cm etcc)
plt.figure(figsize=(10,5))
for col in 'xy':
    plt.hist(data[col], normed=True,alpha=0.7)
```

#### In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable. Kernel density estimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample.

```python
# Plotting Kernel Density Estimation 
plt.figure(figsize=(10,5))
for col in 'xy':
    sns.kdeplot(data[col],shade=True)
```

```python
# And finally Histograms and KDE can be combined using distplot
plt.figure(figsize=(20,5))
for i,col in enumerate('xy'):
    plt.subplot(1,2,i+1)
    sns.distplot(data[col])
```

```python
# Passing 2D data directly using kedplot
plt.figure(figsize=(10,5))
sns.kdeplot(data)
```

```python
# Joint Distribution and marginal distribution
plt.figure(figsize=(10,5))
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
# we can try to plot joint distribution on hexogonally based histogram too by changing kind = hex
plt.figure(figsize=(10,5))
with sns.axes_style('white'):
    sns.jointplot("x", "y",data, kind='hex')
```

## Faceted Histograms 

```python
# Some times the best way to view data is via histograms of subsets so that we can specify which data we are interested in
# in below example from the tips dataset we would like to find out the relation between the tips during lunch and dinner
# for both men and women seperately 
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True, size=4.5)
grid.map(plt.hist, 'tip_pct', bins=np.linspace(0,40,15))
```

## Factor Plots

```python
# Factor plots allow us to view the distribution of a parameter within bins defined by another parameters
with sns.axes_style(style='ticks'):
    g = sns.factorplot("day", "total_bill",'sex',data=tips,kind="box",size=7)
    g.set_axis_labels("Day","Total Bill")
```

## Joint Distributions

```python
# we can use sns.jointplot to show the joint distribution between different datasets, 
# along with the associated marginal distributions:
```

```python
with sns.axes_style('white'):
    sns.jointplot('total_bill', 'tip',data=tips, kind='hex',size=8)
```

```python
#  Linear Regression :) 
with sns.axes_style('white'):
    sns.jointplot('total_bill', 'tip', data=tips, kind='reg',size=8)
```

```python
# Kde joint distributions
with sns.axes_style('white'):
    sns.jointplot('total_bill','tip',data=tips,size=7,kind='kde')
```
